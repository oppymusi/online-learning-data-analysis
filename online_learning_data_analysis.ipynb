## MISSING VALUE

```python
df.isna().sum()
```

```python
(df.isna().sum() / df.shape[0]) * 100
```

```python
#mengatasi missing value pada kolom 'tanggal lahir'
df['tanggal_lahir'] = df['tanggal_lahir'].fillna(df['tanggal_lahir'].mean())
```

```python
#mengatasi missing value pada kolom 'tanggal_menyelesaikan'
df['tanggal_menyelesaikan'] = df['tanggal_menyelesaikan'].fillna(df['tanggal_menyelesaikan'].mean())
```

```python
#mengatasi missing value pada kolom 'skor_akhir'
df['skor_akhir'] = df['skor_akhir'].fillna(df['skor_akhir'].mean())
```

```python
#mengatasi missing value pada 'rating_kelas'
group_means = df.groupby('bidang_studi')['rating_kelas'].transform('mean')
df['rating_kelas'] = df['rating_kelas'].fillna(group_means.fillna(df['rating_kelas'].mean()))
```

```python
#mengubah review kelas berdasarkan rating
df['review_kelas'] = df.apply(lambda row:
  'Sangat baik' if row['rating_kelas'] and row['rating_kelas'] >= 5 else
  'Cukup baik' if 4 <= row['rating_kelas'] < 5 else
  'Baik' if 3 <= row['rating_kelas'] < 4 else
  'Kurang' if 2 <= row['rating_kelas'] < 3 else
  'Sangat kurang' if 0 < row['rating_kelas'] < 2 else
  row['review_kelas'], axis=1
  )
```

### Mengubah data dengan nilai tidak sesuai

```python
#fungsi untuk menghilangkan nilai *
def replace(value):
    return value.replace('*', '') if pd.notnull(value) else value

#mengganti nilai * pada nama__pengguna dan nama_kelas
df['nama_pengguna'] = df['nama_pengguna'].str.upper().apply(replace)
df['nama_kelas'] = df['nama_kelas'].str.upper().apply(replace)
```

```python
def cek_outlier(nama_kolom):
    Q1 = nama_kolom.quantile(0.25)
    Q3 = nama_kolom.quantile(0.75)
    IQR = Q3 - Q1

    #batas
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    #deteksi
    outliers = nama_kolom[(nama_kolom < lower_bound) | (nama_kolom > upper_bound)]

    return outliers

#penggunaan
skor = df['skor_akhir']
rating = df['rating_kelas']
aktivitas = df['total_aktivitas']
selesai = df['aktivitas_selesai']

outliers_skor = cek_outlier(skor)
outliers_rating = cek_outlier(rating)
outliers_aktivitas = cek_outlier(aktivitas)
outliers_selesai = cek_outlier(selesai)
```

```python
def replace_outliers(data, column, outliers):
    median_value = df[column].median()
    df[column].loc[outliers.index] = median_value
    df[column] = round(data[column])

replace_outliers(df, 'skor_akhir', outliers_skor)
replace_outliers(df, 'rating_kelas', outliers_rating)
replace_outliers(df, 'total_aktivitas', outliers_aktivitas)
replace_outliers(df, 'aktivitas_selesai',outliers_selesai)
```
